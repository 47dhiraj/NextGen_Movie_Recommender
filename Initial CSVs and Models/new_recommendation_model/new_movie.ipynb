```python
import pandas as pd

# import joblib
# import feather
# import time
```

```python
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')
```

```python
## This cell is just for understanding purpose

# # merging two seperate dataframes to create a single dataframe

# # ratings = pd.merge(movies, ratings)
# # ratings.head()
```

```python
## merging two dataframes, and after that,
## dropping unnecessary columns(i.e genres & timestamps)
## In Pandas, axis=1 always means, column wise

ratings = pd.merge(movies, ratings).drop(['genres','timestamp'], axis=1)
ratings.head()
```

```python
## Creating a dataframe that looks like pivot table (for shake of easy: just like pivot table in excel)

## index=['userId'] vannale, row index ma 'userID' rahos vanna lai
## columns=['title'] vannale, column  headings ma  movie ko 'title' rahos vanna lai
## values='rating' vannale, values ma chai user le movie lai diyeko 'rating' rahos vanna lai

user_ratings = ratings.pivot_table(index=['userId'], columns=['title'], values='rating')
user_ratings.head()
```

```python
## This cell is just for understanding purpose

# # Removing movies which have less than 10 users who rated it and filling remaining NaN values with 0

## dropna(thresh=10,axis=1) --> yesko kaam vaneko,  jun column (i.e movie) ma 10 vanda kam users le rating deko xa.. testo column/movie lai drop gar vaneko.
## fillna(0) ko kaam vaneko column drop gari sake paxi jati column bachxa .. bacheko column ma pani NaN value tw huna saki halxa.. so tyo NaN value lai 0 le replace garni vaneko. 

## NOTE: jati dherai thresh hold values, teti hamro ML model batw movie completely remove huncha, but ML model file ko size less huncha
## So, thresh hold value set garda ekdam preciouse/accurate vayera set garnu parcha (i.e value low/less vayo vani model ma noise badi huncha & value high/more vayo vani pani hamile banayeko model le thuprai movies lai chindaina)

## In pandas, axis=1 vannale, column wise vanne bujnu parcha (i.e ahile dataframe ko column ma movie cha so, movie vanne bujincha)

# # user_ratings = user_ratings.dropna(thresh=10, axis=1).fillna(0)         
# # user_ratings.head()
```

```python
user_ratings = user_ratings.dropna(thresh=10, axis=1).fillna(0)

user_ratings.head()
```

```python
## This cell is just for Understanding purpose

# # Building our Correlation / Similarity Matrix

## NOTE: pearson correlation method automatically replace 0 rating values by mean rating values, unlike cosine similarity
## so, thats the one reason, pearson correlation method is prefered over cosine similary algorithm
```

```python
item_similarity_df = user_ratings.corr(method='pearson')

item_similarity_df.head()
```

```python
## get_similar_movies() vanni chai method ho which accepts movie name ans user rating as parameters

##  get_similar_movies()  method return similar_score i.e similar movies

def get_similar_movies(movie_name,user_rating):
    similar_score = item_similarity_df[movie_name]*(user_rating-2.5)
    similar_score = similar_score.sort_values(ascending=False)
    # print(similar_score)
    return similar_score                                            
```

```python
## Checking if the reommending movies is already watched or not ?

def check_seen(movie, seen_movies):
    for item in seen_movies:
        if item == movie:
            return True

    return False
```

```python
# scifi_lover = [("The Martian", 5)]
scifi_lover = [("The Martian", 5), ("Edge of Tomorrow",4)]

romantic_lover = [("Titanic",5),("Kung Fu Panda",4)]

drama_crime_lover = [("The Shawshank Redemption", 5)]


# # # creating an empty dataframe (i.e similar_movies)
similar_movies = pd.DataFrame()         



for movie,rating in scifi_lover:
    similar_movies = similar_movies._append(get_similar_movies(movie,rating), ignore_index=True)


# for movie,rating in drama_crime_lover:
#     similar_movies = similar_movies._append(get_similar_movies(movie,rating), ignore_index=True)


# for movie,rating in romantic_lover:
#     similar_movies = similar_movies._append(get_similar_movies(movie,rating), ignore_index=True)



similar_movies.head()
```

```python
all_recommend = similar_movies.sum().sort_values(ascending=False)

all_recommend.head(20)
```

```python
# check_name = ['The Martian']
check_name = ['The Martian', 'Edge of Tomorrow']

# check_name = ['The Shawshank Redemption']

print('\nList of similar movies like',check_name, '\n')


i = 0
for movie, score in all_recommend.items():
    if not check_seen(movie, check_name):
        # print(movie,'\t\t\t\t\t',  score)
        print(movie)

    i = i + 1
    if i >= 36 + len(check_name):
        break
```
